Импортируем библиотеки из 3 практической работы с набором данных pandas и numpy. Импортируем библиотеку seaborn для построения графиков. Импортируем и библиотеку matplotlib.pyplot с упрощенными графиками.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plot
```

Выгружаем очищенный набор данных из файла предыдущей работы vgsales_8.csv.

```python
df = pd.read_csv('vgsales_8.csv')
df
```

countplot() создаёт график, позволяющий продемонстрировать количество наблюдений в наборе данных в каждой категориальной ячейке с помощью столбцов.
Построим график для категориального признака Platform из набора данных.

Описание: по данному графику можно сказать, что наибольшее количество игр было выпущено на платформу DS и PS2. Ближайшим преследователем будет PS3, имеющая примерно 55 выпусков. Худшими стали HG и DC - менее 5

```python
graph = sns.countplot(x="Platform", data = df)
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 
```

barplot() аггрегирует данные по значениям категориальной переменной и применяет определённую функцию к значениям соответсвующих групп цифровой переменной. По умолчанию эта функция — среднее.

Описание: В среднем наибольшее количество мировых продаж видеоигр приходится на GB, второй идёт SNES, третьей DC. Плафтормой с наименьшими продажами является NG.

```python
graph = sns.barplot(x=df["Platform"], y=df["Global_Sales"])
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 
```

boxplot() — ящик с усами или диаграмма размаха. Такой вид диаграммы в удобной форме показывает медиану (или, если нужно, среднее), нижний и верхний квартили, минимальное и максимальное значение выборки и выбросы.

Описание: По графику можно наблюдать, что самыми продаваемыми играми по жанру в среднем являются шутеры (Shooter), вторым идут ролевые игры (Role-Playing), третьими идут гонки (Racing). Самым непопулярным жанром являются адвенчуры (Adventure).

```python
graph = sns.boxplot(x="Genre", y="Global_Sales",data = df,palette='rainbow')
graph.figure.set_figwidth(15)
graph.figure.set_figheight(10) 
```

Тепловая карта (heatmap()) - это метод визуализации данных, который показывает величину явления в виде цвета в двух измерениях. Изменение цвета может быть вызвано оттенком или интенсивностью
Описание: В данном контексте эта диаграмма не является релевантной, так как она больше предназначена для анализа температурных данных. Следовательно, ориентироваться на нее для оценки правильности данных не рекомендуется. Но если все-таки постараться что-то сказать по нему, то зависимость между случайными числами Rang и Global_Sales близка к минимуму из предложенного интервала.

```python
MyHeatMap = sns.heatmap(df.corr(numeric_only=True), vmin=-1, vmax=1) #аргумент annot=True отображает коэффициент корреляции
MyHeatMap.figure.set_figwidth(10)
MyHeatMap.figure.set_figheight(5) 
```

pairplot() - построение попарных взаимосвязей в наборе данных.
По умолчанию эта функция создаст сетку осей таким образом, что каждая числовая переменная в данных будет разделена по осям y в одной строке и по осям x в одном столбце. Диагональные графики обрабатываются по-разному: строится одномерный график распределения, показывающий предельное распределение данных в каждом столбце.
Описание: в данном случае график представляет сравнение всех числовых столбцов между собой. Например, возьмем столбцы Year и Rang. На графике можно увидеть, в каком году наблюдался определенный ранг. Аналогично происходит с остальными числовыми столбцами, которые сравниваются между собой.

```python
graph = sns.pairplot(df)
```

PairGrid() - сетка подзаголовков для построения попарных взаимосвязей в наборе данных.
Этот объект сопоставляет каждую переменную в наборе данных со столбцом и строкой в сетке из нескольких осей. Различные функции построения графиков на уровне возрастов могут быть использованы для построения двумерных графиков в верхнем и нижнем треугольниках, а предельное распределение каждой переменной может быть показано по диагонали.
Описание: данная диаграмма имеет тот же функционал, что и предыдущая, но ее визуальное представление отличается благодаря соединенным кружкам. В результате, здесь также осуществляется сравнение всех числовых столбцов между собой.

```python
graph = sns.PairGrid(df)
graph = graph.map_upper(sns.scatterplot)
graph = graph.map_lower(sns.kdeplot)
graph = graph.map_diag(sns.kdeplot, lw = 2)
```

Кластеризация — задача группировки множества объектов на подмножества (кластеры) таким образом, чтобы объекты из одного кластера были более похожи друг на друга, чем на объекты из других кластеров по какому-либо критерию.
Описание: в данном случае мы начинаем с выбора нескольких столбцов и для упрощения демонстрации рассматриваем только значения, кратные 50.

```python
df_mini = df[["NA_Sales","EU_Sales","JP_Sales"]][::50]
df_mini
```

Это итеративный алгоритм кластеризации, основанный на минимизации суммарных квадратичных отклонений точек подмножества от средних координат этих подмножеств.
 Описание: Начинаем с выбора желаемого количества кластеров - 3 (описываем модель). Затем проводим моделирование, используя метод fit, и предсказываем точки на всем наборе данных с помощью метода predict. Оставшаяся задача - вывести эти предсказания в отдельно созданный столбец.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3)

model.fit(df_mini)

all_predictions = model.predict(df_mini)

df_mini["KMeans"] = all_predictions

df_mini
```

Описание: демонстрация KMeans на самой простой диаграмме,отображающей количество.

```python
sns.countplot(data = df_mini, y = "KMeans")
```

Описание: демонстрация KMeans на pairplot диаграмме,отображающей нахождения подмножеств из нового столбца в стобцах указанных параметре.

```python
sns.pairplot(df_mini,hue="KMeans")
```
